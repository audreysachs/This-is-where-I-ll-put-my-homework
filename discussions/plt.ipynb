```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 100)
#print(x)
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.figure()
plt.plot(x, y)
plt.plot(x, y2)
plt.title("example")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['sine','cosine'], loc="upper right")
plt.show()
```
